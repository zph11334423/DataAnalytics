```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('film.csv',sep="|", quotechar="'")
df1
```

```python
df2 = pd.read_excel('categories.xlsx')
df2
```

```python
merge = pd.merge(df1, df2, left_on='category_id', right_on='category_id')
merge
```

```python
category = merge['category_name'].value_counts()
category
```

```python
category.plot(kind='bar', xlabel='Film Category', ylabel='Number of Films', title='Number of Films by Category')
plt.show()
```

```python
plt.figure(figsize=(20, 7))
duration = merge.groupby('category_name')['rental_duration']
plt.boxplot([group for name, group in duration], labels = duration.groups.keys())
plt.xlabel('Category')
plt.ylabel('Rental Duration')
plt.title('Rental Duration by Category')
plt.show()
```
