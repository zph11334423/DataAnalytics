```python
all_df = glob.glob('*.csv')

info = pd.read_excel('STL_players.xlsx')

name_data = []

joined = [file for file in all_df if 'joined' in file]
```

```python
for file in joined:
    df = pd.read_csv(file)
    name = pd.merge(df, info, left_on = 'toucher', right_on = 'id')
    name = name[['FrameCount', 'Second', 'id', 'name', 'position', 
            '176', '177', '178', '179', '180', '181', 'successful']]
    name_data.append(name)
```

```python
name_data
```

```python
output = 'with_name/'

os.makedirs(output, exist_ok=True)

for name_df, path in zip (name_data, joined):
    new_name = os.path.basename(path).replace('joined.csv', 'name.csv')
    new_path = os.path.join(output, new_name)
    name_df.to_csv(new_path, index = False)
```

```python
combined = pd.concat(name_data, ignore_index = True)
combined
```
